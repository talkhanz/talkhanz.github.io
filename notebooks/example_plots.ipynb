# Example plots

Running this notebook writes plot assets into `docs/assets/plots/`:

- `loss_curve.png` — a static matplotlib figure
- `scaling.html` — a self-contained interactive Plotly fragment

Regenerate with `make plots` from the repo root.

```python
from pathlib import Path

import numpy as np

# Work whether the notebook is run from the repo root or from notebooks/.
root = Path.cwd()
if not (root / "docs").exists():
    root = root.parent
OUT = root / "docs" / "assets" / "plots"
OUT.mkdir(parents=True, exist_ok=True)
rng = np.random.default_rng(0)
print("Writing plots to", OUT)
```

## Static figure (matplotlib → PNG)

```python
import matplotlib

matplotlib.use("Agg")
import matplotlib.pyplot as plt

steps = np.arange(1, 201)
train = 3.0 * np.exp(-steps / 40) + 0.25 + rng.normal(0, 0.02, steps.size)
val = 3.0 * np.exp(-steps / 55) + 0.35 + rng.normal(0, 0.02, steps.size)

fig, ax = plt.subplots(figsize=(7, 4))
ax.plot(steps, train, label="train", lw=2)
ax.plot(steps, val, label="val", lw=2)
ax.set_xlabel("step")
ax.set_ylabel("loss")
ax.set_title("Training vs. validation loss")
ax.legend()
ax.grid(True, alpha=0.3)
fig.savefig(OUT / "loss_curve.png", dpi=150, bbox_inches="tight")
plt.close(fig)
print("wrote", OUT / "loss_curve.png")
```

## Interactive figure (Plotly → inlined HTML fragment)

`full_html=False` writes just a `<div>` + `<script>`; `include_plotlyjs="cdn"`
loads plotly.js from a CDN. The fragment is inlined into the post with a MkDocs
snippet include.

```python
import plotly.graph_objects as go

# A toy neural-scaling-law curve: loss vs. compute for a few model sizes.
compute = np.logspace(0, 4, 50)
fig = go.Figure()
for params, c0 in [("125M", 1.0), ("350M", 0.8), ("1.3B", 0.65)]:
    loss = c0 * compute ** (-0.07) + 1.8
    fig.add_trace(
        go.Scatter(x=compute, y=loss, mode="lines", name=params)
    )
fig.update_layout(
    title="Toy scaling law: loss vs. compute",
    xaxis=dict(title="compute (arb. units)", type="log"),
    yaxis=dict(title="loss"),
    template="plotly_white",
    legend_title="params",
    height=460,
    margin=dict(l=60, r=20, t=60, b=50),
)
fig.write_html(
    OUT / "scaling.html",
    include_plotlyjs="cdn",
    full_html=False,
    div_id="scaling-plot",
)
print("wrote", OUT / "scaling.html")
```
